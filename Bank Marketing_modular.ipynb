```python
def hist(col):
    plt.figure(figsize=(10, 6))
    sns.histplot(data[data['y'] == 0][col], color='blue', label='Not subscribed', kde=True)
    sns.histplot(data[data['y'] == 1][col], color='orange', label='Subscribed', kde=True)
    plt.legend()
    plt.show()
```

```python
def bar(col,col1='y'):
    
    import matplotlib.pyplot as plt
    import seaborn as sns
    
    # Assuming 'data' is your DataFrame
    job_subscriptions = data.groupby(col)[col1].value_counts(normalize=True).unstack()
    
    # Plotting
    plt.figure(figsize=(12, 6))
    sns.barplot(data=job_subscriptions.reset_index(), x=col, y=0, color='green', label='Not Subscribed')
    sns.barplot(data=job_subscriptions.reset_index(), x=col, y=1, color='blue', label='Subscribed')
    
    # Adding annotations
    for i, (not_subscribed, subscribed) in enumerate(zip(job_subscriptions[0], job_subscriptions[1])):
        plt.text(i, not_subscribed / 2, f'{not_subscribed:.2%}', va='center', ha='center', color='white', fontsize=10, weight='bold')
        plt.text(i, not_subscribed + subscribed / 2, f'{subscribed:.2%}', va='center', ha='center', color='white', fontsize=10, weight='bold')
    
    # Adding labels and title
    plt.xlabel(f'{col}')
    plt.ylabel('Percentage')
    plt.title(f'Subscription Status by {col}')
    plt.legend(title='Subscription Status')
    
    # Display the plot
    plt.xticks(rotation=45, ha='right')
    plt.show()
```

```python
def chi(col1, col2):
    import pandas as pd
    from scipy.stats import chi2_contingency
    
    # Assuming 'data' is your DataFrame
    job_education_contingency = pd.crosstab(data[col1], data[col2])
    
    # Performing Chi-Square test
    chi2, p, _, _ = chi2_contingency(job_education_contingency)
    
    # Display results
    # print(f"Chi-Square Value: {chi2}")
    # print(f"P-Value: {p}")
    
    # Interpret the results
    alpha = 0.05
    if p < alpha:
        return 1
        # print(f"There is a significant association between {col1} and {col2}.")
    else:
        return 0
        # print(f"There is no significant association between {col1} and {col2}.")
```

```python
# Assuming 'data' is your DataFrame
numerical_columns = df.select_dtypes(include='number').columns
for column in numerical_columns:
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    outliers = (df[column] < Q1 - 1.5 * IQR) | (df[column] > Q3 + 1.5 * IQR)
    if any(outliers):
        print(f'{column}: has outliers')
        # print(f'Outliers in {column}: {df[outliers][column].tolist()}')
        
```

### MLFlow

```python
def create_experiment(experiment_name, run_name, run_metrics, model, confusion_matrix_path=None, roc_auc_curve_path=None
                     , run_params=None):
    import mlflow
    from mlflow.tracking import MlflowClient

    mlflow.set_experiment(experiment_name)

    with mlflow.start_run() as run:
        run_id = run.info.run_id
        
        if run_params:
            for param in run_params:
                mlflow.log_param(param,run_params[param])

        for metric in run_metrics:
            mlflow.log_metric(metric,run_metrics[metric])

        mlflow.sklearn.log_model(model, "model")

        if confusion_matrix_path:
            mlflow.log_artifact(confusion_matrix_path, "confusion_matrix")

        if roc_auc_curve_path:
            mlflow.log_artifact(roc_auc_curve_path, "roc_auc_curve")
        
        mlflow.set_tags({'tag1':'Logisticregression','tag2':'classification'})
        print(run_name, "is logged to experiment-", experiment_name)
        
        client = MlflowClient()
        client.update_run(run_id,name=run_name) #for changing the 'run name'
```
